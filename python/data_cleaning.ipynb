```python
import pandas as pd
import os

df = pd.read_csv(r"C:\Users\deepj\Documents\customer-profitability-analysis\data\raw\online_retail.csv", encoding="latin1")

print(df.shape)
df.head()
```

```python
print(df.isnull().sum())
print(df.dtypes)
```

```python
df = df[~df["InvoiceNo"].astype(str).str.startswith("C")]
df = df.dropna(subset=["CustomerID"])
df = df[df["Quantity"] > 0]
df = df[df["UnitPrice"] > 0]
print("cleaned shape:", df.shape)
```

```python
df["revenue"] = (df["Quantity"] * df["UnitPrice"]).round(2)
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
df["year"] = df["InvoiceDate"].dt.year
df["month"] = df["InvoiceDate"].dt.month
df["year_month"] = df["InvoiceDate"].dt.to_period("M").astype(str)
df.head()
```

```python
os.makedirs(r"C:\Users\deepj\Documents\customer-profitability-analysis\data\processed", exist_ok=True)
df.to_csv(r"C:\Users\deepj\Documents\customer-profitability-analysis\data\processed\retail_cleaned.csv", index=False)
print("saved:", df.shape)
```
